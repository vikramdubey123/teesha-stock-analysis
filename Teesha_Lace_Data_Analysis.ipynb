```python
#Import libraries and read the uploaded Excel file
import pandas as pd

# Replace this with your exact uploaded file name
file_name = "Finish Stock Summary11-07-2025 10.42.43.xlsx"

# Read the file
df = pd.read_excel(file_name)

# Show first few rows
df.head()
```

```python
# Data Cleaning - replace NaNs with 0
df_cleaned = df.fillna(0)

# Make sure 'GrandTotal' column is consistent
df_cleaned.columns = df_cleaned.columns.str.strip()
df_cleaned = df_cleaned.rename(columns={'GrandTotal': 'Total_Pcs'})

# Preview cleaned data
df_cleaned.head()
```

Check dataset size and structure


```python
print("Rows:", df_cleaned.shape[0])
print("Columns:", df_cleaned.shape[1])
df_cleaned.info()
```

View first few rows

```python
df_cleaned.head(5)
```

List of all color columns (excluding D_NO and Total_Pcs)

```python
color_columns = df_cleaned.columns[1:-1]
print("All Color Columns:", list(color_columns))
```

Count how many colors are non-zero in each design

```python
df_cleaned['Color_Count'] = df_cleaned[color_columns].gt(0).sum(axis=1)
df_cleaned[['D_NO', 'Color_Count', 'Total_Pcs']].head()
```

Sort and show top 5 stocked designs

```python
top_5 = df_cleaned.sort_values(by='Total_Pcs', ascending=False).head(5)
top_5[['D_NO', 'Total_Pcs']]
```

Sort and show bottom 5 stocked designs

```python
bottom_5 = df_cleaned.sort_values(by='Total_Pcs', ascending=True).head(5)
bottom_5[['D_NO', 'Total_Pcs']]
```

Bar chart for top 10 designs by total stock

```python
import seaborn as sns
import matplotlib.pyplot as plt

top_10 = df_cleaned.sort_values(by='Total_Pcs', ascending=False).head(10)
sns.barplot(data=top_10, x='D_NO', y='Total_Pcs', palette='viridis')
plt.title("Top 10 Designs by Total Stock")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Total stock per color (overall, not per design)

```python
color_totals = df_cleaned[color_columns].sum().sort_values(ascending=False)
color_totals.head(10)
```

Bar plot of total stock per color

```python
color_totals.plot(kind='bar', color='teal')
plt.title("Total Stock by Color (All Designs)")
plt.ylabel("Pieces")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Pie chart for color distribution in one design

```python
design_name = "3-SAMOSA"  # Change this to any design
row = df_cleaned[df_cleaned['D_NO'] == design_name][color_columns].iloc[0]
row = row[row > 0]  # Only non-zero colors

row.plot(kind='pie', autopct='%1.1f%%', startangle=140, figsize=(6, 6))
plt.title(f"Color Distribution in {design_name}")
plt.ylabel("")
plt.tight_layout()
plt.show()
```

Color-wise average stock across all designs

```python
df_cleaned[color_columns].mean().sort_values(ascending=False).head(10)
```

Designs vs Colors (stock level)

```python
heatmap_data = df_cleaned.set_index('D_NO')[color_columns]
sns.heatmap(heatmap_data, cmap='YlGnBu', linewidths=0.2)
plt.title("Design vs Color Stock Heatmap")
plt.xlabel("Colors")
plt.ylabel("Designs")
plt.tight_layout()
plt.show()
```

Correlation between color stocks

```python
sns.heatmap(df_cleaned[color_columns].corr(), annot=False, cmap='coolwarm')
plt.title("Color Correlation")
plt.tight_layout()
plt.show()
```

Top 5 most frequent colors per design

```python
df_cleaned['Top_Color'] = df_cleaned[color_columns].idxmax(axis=1)
df_cleaned[['D_NO', 'Top_Color']].head()
```

Count how many times each color is the top in designs

```python
df_cleaned['Top_Color'].value_counts().head(10)
```

Filter designs where any color has more than 500 pcs

```python
filtered = df_cleaned[(df_cleaned[color_columns] > 500).any(axis=1)]
filtered[['D_NO', 'Total_Pcs']].head()
```

Create a summary DataFrame (design, total_pcs, color_count, top_color)

```python
summary_df = df_cleaned[['D_NO', 'Design_Name', 'Total_Pcs', 'Color_Count', 'Top_Color']]
summary_df.head(10)
```
